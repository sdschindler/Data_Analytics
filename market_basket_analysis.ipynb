```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

df = pd.read_csv('teleco_market_basket.csv')
df.info()
```

```python
df.shape
```

```python
df.dropna(axis=0, how='all',inplace=True)
#7501 X 20 is what it should be
df.shape
```

```python
# create list of lists
trans = []
for i in range(0,7501):
    trans.append([str(df.values[i,j]) for j in range(0,20)])
```

```python
TE = TransactionEncoder()
array = TE.fit(trans).transform(trans)

```

```python
cleaned_df = pd.DataFrame(array,columns=TE.columns_)
cleaned_df
```

```python
for col in cleaned_df.columns:
    print(col)
```

```python
df_cleaned = cleaned_df.drop(['nan'],axis =1)
for col in df_cleaned.columns:
    print(col)
```

```python
df_cleaned.shape
```

```python
df_cleaned.to_csv("clean_d212_MBA.csv")
```

```python
a_rules = apriori(df_cleaned, min_support = 0.05, use_colnames = True)
a_rules.nlargest(3, ['support'])
```

```python
assoc_rules = association_rules(a_rules, metric = 'lift', min_threshold = 1)
assoc_rules
```

```python
assoc_rules[ (assoc_rules['lift'] >= 1.3) & 
              (assoc_rules['confidence'] >= .25) & (assoc_rules['support'] >= .05)  ]
```
